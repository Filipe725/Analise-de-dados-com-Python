```python
import numpy as np
```

```python
alunos = np.array(['joão', 'marcos', 'maria'])
notas_matematica = np.array([7, 6, 5])
notas_fisica = np.array([10, 9, 8])
```

```python
df = np.column_stack((alunos, notas_matematica, notas_fisica))
```

```python
media = np.column_stack((notas_fisica, notas_matematica))
medias = np.mean(media, axis=1)
medias_maioroes_que_7 = medias > 7
df = np.column_stack((alunos, medias))
aprovados = df[medias_maioroes_que_7]
aprovados
```

```python
notas = np.vstack((alunos, notas_fisica, notas_matematica))
notas
```
